```python
import cv2
```

```python
# Load the sample image
image = cv2.imread('1691389133304.png')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
import numpy as np

# Define a 3x3 convolution kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Apply convolution operation on the image
convolved_image = cv2.filter2D(gray_image, -1, kernel)
```

```python
# Apply min pooling
min_pooling_image = cv2.erode(gray_image, kernel=(3, 3), iterations=1)

# Apply max pooling
max_pooling_image = cv2.dilate(gray_image, kernel=(3, 3), iterations=1)
```

```python
# Apply padding to the image
padded_image = np.pad(gray_image, ((1, 1), (1, 1)), mode='constant')
```

```python
import numpy as np

# Define a stride of 4
stride = 4

# Apply convolution with stride of 4
convolved_stride_image = cv2.filter2D(gray_image[::stride, ::stride], -1, kernel)
```

```python
import matplotlib.pyplot as plt

# Plot the images
plt.figure(figsize=(10, 10))

plt.subplot(3, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(3, 2, 3)
plt.imshow(convolved_image, cmap='gray')
plt.title('Convolved Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(min_pooling_image, cmap='gray')
plt.title('Min Pooling Image')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(max_pooling_image, cmap='gray')
plt.title('Max Pooling Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.imshow(convolved_stride_image, cmap='gray')
plt.title('Convolved Image with Stride 4')
plt.axis('off')

plt.show()
```
