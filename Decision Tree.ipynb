# Labels to numbers 

```python
#Ml Model does not work on labels so we need to convert the labels to numbers.
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()## We create the label encoder for each column

#Create Label Encoder for other two columns as well
"Your Code Here"
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
data_1['company_n'] = le_company.fit_transform(data_1['company']) # It will transform the company column to new column company_n
# You have to transform othe column 
"Your Code Here"
data_1['job_n'] = le_job.fit_transform(data_1['job'])
data_1['degree_n'] = le_degree.fit_transform(data_1['degree'])
```
